```python
import pandas as pd
```

```python
df=pd.read_csv("lung_cancer_examples.csv")
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc=LabelEncoder()
```

```python
enc.fit(df.Name)
```

```python
df.Name=enc.transform(df.Name)
```

```python
df
```

```python
x=df.iloc[:,[2,3,4,5]].values
```

```python
y=df.iloc[:,6].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std=StandardScaler()
x_train=std.fit_transform(x_train)
x_test=std.fit_transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
p=model.score(x_test,y_test)*100
p
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
q=model.score(x_test,y_test)*100
q
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor(n_estimators='warn')
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict([[3,5,4,1]])
y_pred
```

```python
r=model.score(x_test,y_test)*100
r
```

```python
import matplotlib.pyplot as plt
```

```python
x1=[1,2,3]
y1=[p,q,r]

plt.xlabel("Algorithms")
plt.ylabel("Accuracy")
plt.title("Lung Cancer Graph")
plt.bar(x1,y1,width=0.5,color='g',label="Decision Tree ")
plt.bar(x1,y1,width=0.5,color='r',label="Logistic Regression")
plt.bar(x1,y1,width=0.5,color='b',label="Random Forest Regressor")
plt.bar(x1,y1,width=0.5,color=['g','r','b'])
plt.legend()
```
